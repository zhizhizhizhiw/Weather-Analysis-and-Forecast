```python
from pmdarima import auto_arima
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import datetime as dt
```

```python
weather = pd.read_csv("cleaned_weather.csv")
mean_temp = (weather["temp_low"] + weather["temp_high"]) / 2
mean_temp = mean_temp.dropna()
```

```python
mean_temp.plot(x="date", y="mean_temp", title="dalian_weather_perday", figsize=(18,6))
plt.xlabel("Date")
plt.ylabel("Mean Temperature")
plt.grid()
plt.show()
```

```python
train_data = mean_temp[:1200]
test_data = mean_temp[1200:]
```

```python
auto_model = auto_arima(mean_temp, seasonal=True, m=12, trace=True, max_p=3, max_q=3, max_P=2, max_Q=2, max_D=1, n_jobs=1, suppress_warnings=True, stepwise=True)
print(auto_model.summary())
```

```python
best_order = auto_model.order
print(best_order)
```

```python
sarima_model = SARIMAX(mean_temp, order=best_order)
sarima_model_fit = sarima_model.fit()
```

```python
x_train = range(len(train_data))
x_test = range(len(train_data), len(test_data)+len(train_data))

plt.figure(figsize=(18, 6))

plt.plot(x_train, train_data, label='train_data')
plt.plot(x_test, test_data, label='test_data', color="red")
plt.plot(x_test, sarima_model_fit.fittedvalues[-len(test_data):], label='mean_temp_fit', color="orange")

plt.xlabel('Time(day)')
plt.ylabel('mean_temp(℃)')
plt.title("dalian_weather")

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

mse = mean_squared_error(mean_temp, sarima_model_fit.fittedvalues)
mape = mean_absolute_percentage_error(mean_temp, sarima_model_fit.fittedvalues)

print(f"Hybrid模型 MSE: {mse:.4f}")
print(f"Hybrid模型 MAPE: {mape:.4%}")
```
